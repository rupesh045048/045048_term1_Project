```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Lists to store data
Product_name = []
Current_price = []
MRP = []
Disc_rate = []
Reviews = []
N_Rate_Rev = [] #no. of rating and reviews 
Description = []

for i in range(1, 10):
    URL = 'https://www.flipkart.com/search?q=mobile%20under%20100000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=' + str(i)

    # Headers for request
    HEADERS = ({
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36 OPR/100.0.0.0',
        'Accept-Language': 'en-US, en;q=0.5'})

    # HTTP Request
    webpage = requests.get(URL, headers=HEADERS)

    # Soup Object containing all data
    soup = BeautifulSoup(webpage.content, "html.parser")

    box = soup.find_all("div", class_="_2kHMtA")

    for element in box:
        # Find names
        names = element.find_all("div", class_="_4rR01T")
        for name in names:
            Product_name.append(name.text)

        # Find Current price
        prices = element.find_all("div", class_="_30jeq3 _1_WHN1")
        for price in prices:
            Current_price.append(price.text)

        # Find MRP
        mrp = element.find_all("div", class_="_3I9_wc _27UcVY")
        for m in mrp:
            MRP.append(m.text)

        # Find Disc %age
        disc = element.find_all("div", class_="_3Ay6Sb")
        for d in disc:
            Disc_rate.append(d.text)

        # Find reviews
        reviews = element.find_all("div", class_="_3LWZlK")
        for review in reviews:
            Reviews.append(review.text)

            
        # Find no. of ratings and reviews 
        number = element.find_all("span", class_="_2_R_DZ")
        for n in number:
            N_Rate_Rev.append(n.text)
            
        # Find descriptions
        desc = element.find_all("ul", class_="_1xgFaf")
        for descs in desc:
            Description.append(descs.text)

# All lists should have the same length
max_length = max(len(Product_name), len(Current_price), len(Reviews), len(Description), len(MRP), len(Disc_rate))

# Pad shorter lists with empty strings to match the maximum length
Product_name += [''] * (max_length - len(Product_name))
Current_price += [''] * (max_length - len(Current_price))
MRP += [''] * (max_length - len(MRP))
Disc_rate += [''] * (max_length - len(Disc_rate))
Reviews += [''] * (max_length - len(Reviews))
N_Rate_Rev += [''] * (max_length - len(N_Rate_Rev))
Description += [''] * (max_length - len(Description))

# Create the DataFrame
df = pd.DataFrame({"Product Name": Product_name, "Current Price": Current_price, "MRP": MRP, "Discount %age": Disc_rate, "Reviews": Reviews,"Number Rate & Review" : N_Rate_Rev ,"Description": Description})

# Print the DataFrame
df
```

```python
df.to_csv("FlipScrapedData.csv")
```
